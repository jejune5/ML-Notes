```python
import numpy as np


class BuildMyW2VFromLargeW2V:
    def __init__(self, corpus: list, large_embdding_path: str):
        self.corpus = corpus
        self.large_embdding_loader = LoadWord2VecEmbedding(large_embdding_path)
        # self.large_embdding_loader = embed
        self.large_w2v = self.large_embdding_loader.get_model()
        self.vector_size = self.large_w2v.vector_size

        self.token2vector = {}
        for token in self.corpus:
            if token in self.large_embdding_loader.get_vocab():
                self.token2vector[token] = self.large_embdding_loader.word2vec(word=token)[token]
            else:
                self.token2vector[token] = np.zeros(self.vector_size)

    def get_vocab(self) -> dict:
        vocab = {token: i for i, token in enumerate(self.corpus)}
        return vocab

    def get_vectors(self) -> np.ndarray:
        vectors = np.conj(list(self.token2vector.values()))
        return vectors

    def token2vec(self, token=None):
        if token and token in self.token2vector:
            return {token: self.token2vector[token]}
        else:
            return self.token2vector
```
